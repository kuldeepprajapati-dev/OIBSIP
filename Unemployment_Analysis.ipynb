```python
#importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the unemployment dataset
data = pd.read_csv("/Unemployment in India.csv")
```

```python
#basic information about the data
print(data.info())
```

```python
#getting first 10 rows of dataset
data.head(10)
```

```python
data.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
```

```python
data.head()
```

```python
# Checking for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
data = data.dropna()
```

```python
print(data.isnull().sum())
```

```python
#getting summarized statisctics of data
print(data.describe())
```

```python
#Grouping  data by region and calculating the average unemployment rate for each region
average_unemployment_by_region = data.groupby('Region')["Estimated Unemployment Rate"].mean()
print('Average Unemployment Rate by Region:\n', average_unemployment_by_region)
```

# **CREATING VISUALISATIONS**

```python
#creating bar graph for Average Unemployment Rate v/s Region
plt.figure(figsize=(10, 6))
average_unemployment_by_region.plot(kind='bar', color='orange')
plt.title('Average Unemployment Rate by Each Region', fontsize=18)
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.show()
```

```python
#bar graph for total rural and urban Unemployment
plt.title('Rural Unemployment & Urban UnEmployment', fontsize=16)
sns.countplot(x=data['Area'],palette = "Set1")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

group = data.groupby(["Region"])["Estimated Unemployment Rate"].aggregate("max").reset_index()

plt.figure(figsize=(12, 8))
sns.pointplot(x=group['Region'].values, y=group['Estimated Unemployment Rate'].values, color='blue')
plt.scatter(x=group['Region'].values, y=group['Estimated Unemployment Rate'].values, color='blue', alpha=0.8)  # Set the alpha value for transparency
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by Region", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
import plotly.express as px

unemployment = data[["Region", "Area", "Estimated Unemployment Rate"]]

figure = px.sunburst(
    unemployment,
    path=["Region", "Area"],
    values="Estimated Unemployment Rate",
    color="Estimated Unemployment Rate",
    color_continuous_scale="Portland",  # Use the "Portland" color scale
    width=1000,
    height=1000,
    title="Unemployment Rate in India by Area",
)

# Add labels to improve readability
figure.update_traces(textinfo='label+percent parent')

figure.show()
```
